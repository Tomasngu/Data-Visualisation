```python
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec

# Taken from 4th notebook of BI-VIZ course
def univariate_cont_ds(df, column, title):
    stats_df = pd.DataFrame(np.array([
        ["Minimal value", df[column].min()], 
        ["Maximal value", df[column].max()], 
        ["Range", df[column].max() - df[column].min()],
        ["Average", df[column].mean()],
        ["Median", df[column].median()],
        ["Variance", df[column].var()]
    ]), columns=['Title', 'Value'])
    
    display(stats_df)
    
    # figure layout initialization
    fig = plt.figure(figsize=(10,10), constrained_layout=True)
    fig.suptitle(title, fontsize = 18)
    spec = gridspec.GridSpec(ncols=2, nrows=2, figure=fig)
    ax1 = fig.add_subplot(spec[0, :]) 
    ax2 = fig.add_subplot(spec[1, 0]) 
    ax3 = fig.add_subplot(spec[1, 1])
    
    sns.histplot(data=df, x=column, kde=True, ax=ax1)
    
    # x and y - rotation of plots
    sns.boxplot(data=df, y=column, ax=ax2, medianprops={"color": "pink"})
    sns.violinplot(data=df, y=column, ax=ax3)
```

Vid√≠me, ≈æe zdaleka nejhranƒõj≈°√≠ je form√°t 10+0. Vid√≠me, ≈æe nejhranƒõj≈°√≠ maj√≠ poƒç√°teƒçn√≠ ƒças pohybuj√≠c√≠ se od 5 do 15 minut. 
Nyn√≠ by n√°s mohlo zaj√≠mat nejhranƒõj≈°√≠ form√°ty podle poƒç√°teƒçn√≠ch ƒças≈Ø a jejich rozdƒõlen√≠ do FIDE ƒçasov√Ωch form√°t≈Ø.

**Hern√≠ form√°ty**
- `Bullet`: Do 3 minut
- `Blitz`: 3-10 minut
- `Rapid`: 10-60 minut
- `Classical`: 60 a v√≠ce minut

```python
fo = orig_data[['increment_code']].copy()
def convert(code):
    return int(code.split('+')[0])
def get_format(init):
    if init < 3:
        return 'Bullet'
    elif init < 10:
        return 'Blitz'
    elif init < 60:
        return 'Rapid'
    else:
        return 'Classical'
    
fo['initial_time'] = fo['increment_code'].apply(convert)
fo['format'] = fo['initial_time'].apply(get_format)
```

V r√°mci v√Ωbƒõru modelu m≈Ø≈æeme experimentovat i s r≈Øzn√Ωmi dal≈°√≠mi zp≈Øsoby v√Ωbƒõru a transformace p≈ô√≠znak≈Ø. Tyto √∫pravy budeme pova≈æovat za souƒç√°st modelu a zabal√≠me je spoleƒçnƒõ s natr√©novan√Ωm modelem do Pipeline.

Definujme nyn√≠ nƒõjak√© pomocn√© funkce pro √∫pravu p≈ô√≠znak≈Ø.

```python
def prep_add_game_len_secs(df):
    df = df.copy()
    df["game_len_s"] = df["last_move_at"] - df["created_at"]
    return df

STUFF_TO_DROP = [
    "victory_status",
    "last_move_at",
    "created_at",
    "moves",
    "increment_code",
    "white_id",
    "black_id",
    "opening_name",
    "opening_eco",
]
def mk_prep_dropper(keep=set()):
    def prep_drop(df):
        df = df.copy()
        for col in STUFF_TO_DROP:
            if col not in keep:
                df.drop(col, inplace=True, axis=1)
        return df
    return prep_drop

def prep_one_hot_categories(df: pd.DataFrame):
    for col in df.columns:
        if df[col].dtype == "category":
            df = df.join(pd.get_dummies(df[col])).drop(col, axis=1)
    return df
```

## Tr√©nov√°n√≠ modelu

Nyn√≠ definujme funkci, kter√° dostane parametrem funkce pro √∫pravu p≈ô√≠znak≈Ø a pokus√≠ se naj√≠t optim√°ln√≠ rozhodovac√≠ strom. T√≠m je m√≠nƒõno to, ≈æe bude zkou≈°et r≈Øzn√© hodnoty hyperparametr≈Ø, a vybere ten rozhodovac√≠ strom, kter√Ω je po natr√©nov√°n√≠ na testovac√≠ch datech nej√∫spƒõ≈°nƒõj≈°√≠ na validaƒçn√≠ datech. Hyperparametry, kter√© budeme ladit, bude `criterion`, aneb krit√©rium, pomoc√≠ nƒõho≈æ se mƒõ≈ô√≠ informaƒçn√≠ zisk, a `max_depth`, tedy hloubka stromu.

√öspƒõ≈°nost modelu budeme mƒõ≈ôit plochou pod ROC k≈ôivkou. ROC k≈ôivka je velmi u≈æiteƒçn√° metoda pro vizualizaci toho, jak dob≈ôe funguj√≠ modely pro klasifikaci. ROC k≈ôivku vykresl√≠me n√≠≈æe. Nyn√≠ budeme pouze vyu≈æ√≠vat obsah plochy pod n√≠, co≈æ je ƒç√≠slo, pro nƒõj≈æ plat√≠, ≈æe ƒç√≠m je vy≈°≈°√≠, t√≠m lep≈°√≠ je kvalita modelu. Hodnota 0.5 odpov√≠d√° modelu, kter√Ω ƒçin√≠ n√°hodn√© predikce, a hodnota 1 odpov√≠d√° dokonal√©mu modelu, kter√Ω m√° v≈ædy pravdu. Lze proto oƒçek√°vat, ≈æe hodnota pro n√°≈° model se bude pohybovat v rozmez√≠ 0.5-1.0. 

Funkce tak√© zobraz√≠ z√°vislost √∫spƒõ≈°nosti na validaƒçn√≠ch datech v≈Øƒçi hloubce rozhodovac√≠ho stromu.

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import FunctionTransformer
from sklearn.metrics import roc_auc_score

def find_decision_tree(prep_fns):
    pl = Pipeline(
        [
            (f"fn{i}", FunctionTransformer(fn))
            for i, fn in enumerate(prep_fns)
        ] + [
            ("clf", DecisionTreeClassifier(
                criterion="entropy",
                max_depth=10,
                max_features=None,
                random_state=SEED,
            ))
        ]
    )

    best_score = None
    best_params = None
    
    scores = { "entropy": [], "gini": [] }
    max_depths = np.arange(1, 16)
    for max_depth in max_depths:
        for criterion in ("entropy", "gini"):
            candidate_params = {
                "clf__max_depth": max_depth,
                "clf__criterion": criterion,
            }

            pl.set_params(**candidate_params)
            pl.fit(train_data, target_data.loc[train_data.index])

            proba = pl.predict_proba(val_data)
            candidate_score = roc_auc_score(
                target_data.loc[val_data.index],
                proba if proba.shape[1] > 2 else proba[:, 1],
                multi_class="ovo"
            )

            scores[criterion].append(candidate_score)
            if best_score is None or candidate_score > best_score:
                best_score = candidate_score
                best_params = candidate_params

    fig = plt.figure(figsize=(10,8))
    ax = fig.add_subplot()
    ax.plot(max_depths, scores["entropy"], label="entropy")
    ax.plot(max_depths, scores["gini"], label="gini")
    ax.set_xlabel("max_depth")
    ax.set_ylabel("roc_auc_score")
    ax.legend()
    fig.show()

    pl.set_params(**best_params)
    pl.fit(train_data, target_data.loc[train_data.index])
    return best_score, pl
```

Zavolejme nyn√≠ tuto funkci a nakresleme v√Ωsledn√Ω rozhodovac√≠ strom. Zkusme zahodit v≈°echny p≈ô√≠znaky uveden√© v seznamu `STUFF_TO_DROP` kromƒõ `opening_eco`.

```python
decision_tree_score, decision_tree_pipeline = find_decision_tree([
    prep_add_game_len_secs,
    mk_prep_dropper({"opening_eco"}), # drop all columns from STUFF_TO_DROP except opening_eco
    prep_one_hot_categories,
])

decision_tree: DecisionTreeClassifier = decision_tree_pipeline.steps[-1][1]
print(f"score: {decision_tree_score:.3f}, depth: {decision_tree.max_depth}, criterion: {decision_tree.criterion}, max_features: {decision_tree.max_features}")
```

Jak vidno, zv√≠tƒõzil rozhodovac√≠ strom s hloubkou 7 vyu≈æ√≠vaj√≠c√≠ entropii pro mƒõ≈ôen√≠ informaƒçn√≠ho zisku. Na vizualizaci v√Ω≈°e zn√°zor≈àujeme z√°vislost plochy pod ROC k≈ôivkou (metrika, kterou vyu≈æ√≠v√°me pro mƒõ≈ôen√≠ √∫spƒõ≈°nosti modelu) na hloubce stromu, a to zvl√°≈°≈• pro stromy natr√©novan√© pomoc√≠ Entropie a stromy natr√©novan√© pomoc√≠ Gini indexu. Entropie a Gini index z≈ôejme p≈ôi bin√°rn√≠ klasifikaci vedou na velmi podobn√© v√Ωsledky (ale p≈ôedt√≠m jsme zkou≈°eli tern√°rn√≠ klasifikaci a tam Entropie dosahovala viditelnƒõ lep≈°√≠ch √∫spƒõ≈°nost√≠, ne≈æli Gini index). D√°le pozorujeme n√°sleduj√≠c√≠ trend: √∫spƒõ≈°nost modelu s rostouc√≠ hloubkou stromu nejd≈ô√≠ve roste, a pot√© kles√°. Kles√°n√≠ je zp≈Øsobeno p≈ôeuƒçen√≠m - hlub≈°√≠ stromy se v√≠ce adaptuj√≠ na tr√©novac√≠ mno≈æinu na √∫kor toho, jak jsou schopni zgeneralizovat svoje znalosti na validaƒçn√≠ mno≈æinu. Vƒõ≈ô√≠me proto, ≈æe toto kles√°n√≠ pokraƒçuje do nekoneƒçna, a pro hloubku 7 skuteƒçnƒõ nast√°v√° glob√°ln√≠ maximum pro Entropii.

Nyn√≠ m≈Ø≈æeme strom i nakreslit pomoc√≠ funkce `plot_tree`, ale v z√°jmu p≈ôehlednosti zobraz√≠me jen prvn√≠ t≈ôi hladiny.

```python
from sklearn.tree import plot_tree

fig = plt.figure(figsize=(16,16))
_ = plot_tree(decision_tree, ax=fig.add_subplot(), max_depth=2, fontsize=12, feature_names=decision_tree.feature_names_in_)
fig.show()
```

Nyn√≠ zkus√≠me probl√©m ≈ôe≈°it pomoc√≠ n√°hodn√Ωch les≈Ø. Hyperparametry, kter√© budeme ladit, budou:
- `n_estimators`, tedy poƒçet strom≈Ø v n√°hodn√©m lese.
- `max_depth`, tedy hloubka jednotliv√Ωch strom≈Ø v lese
- `max_samples`, tedy velikost n√°hodn√©ho v√Ωbƒõru z p≈Øvodn√≠ch dat, na kter√©m bude ka≈æd√Ω strom natr√©nov√°n (ka≈æd√Ω strom bude natr√©nov√°n na jin√©m n√°hodn√©m v√Ωbƒõru, v tom je kouzlo n√°hodn√©ho lesa).
  
Parametr `criterion` nastav√≠me na `"entropy"` a parametru `max_features` ponech√°me jeho p≈Øvodn√≠ hodnotu `"sqrt"`, kter√° by mƒõla d√°vat dobr√© v√Ωsledky.

Funkce `find_random_forest` opƒõt vykresl√≠ vizualizaci √∫spƒõ≈°nosti modelu v z√°vislosti na hyperparametrech, tentokr√°t ov≈°em bude zkouman√Ω prostor hyperparametr≈Ø tvaru 5x5x5, tak≈æe p≈ôi vizualizaci pou≈æijeme 5 heatmap.

```python
from sklearn.ensemble import RandomForestClassifier

def find_random_forest(prep_fns, **kwargs):
    pl = Pipeline(
        [
            (f"fn{i}", FunctionTransformer(fn))
            for i, fn in enumerate(prep_fns)
        ] + [
            ("clf", RandomForestClassifier(**{
                **{
                    "criterion": "entropy",
                    "max_depth": 1,
                    "max_samples": 0.1,
                    "random_state": SEED,
                    "n_estimators": 1000,
                },
                **kwargs
            }))
        ]
    )

    best_score = None
    best_params = None

    n_estimatorses = np.arange(100, 501, 100)
    max_depths = np.arange(1, 6)
    max_sampleses = np.linspace(0.2, 1, 5)
    scores = np.zeros((
        len(n_estimatorses),
        len(max_depths),
        len(max_sampleses)
    ))
    
    for i, n_estimators in enumerate(n_estimatorses):
        print(f"level {i + 1}/{len(n_estimatorses)}: ", end="")
        for j, max_depth in enumerate(max_depths):
            print("|", end="")
            for k, max_samples in enumerate(max_sampleses):
                print(".", end="")
                candidate_params = {
                    "clf__n_estimators": n_estimators,
                    "clf__max_depth": max_depth,
                    "clf__max_samples": max_samples,
                }

                pl.set_params(**candidate_params)
                pl.fit(train_data, target_data.loc[train_data.index])

                proba = pl.predict_proba(val_data)
                candidate_score = roc_auc_score(
                    target_data.loc[val_data.index],
                    proba if proba.shape[1] > 2 else proba[:, 1],
                    multi_class="ovo"
                )

                scores[i, j, k] = candidate_score
                if best_score is None or candidate_score > best_score:
                    best_score = candidate_score
                    best_params = candidate_params
        print()

    fig = plt.figure(figsize=(8,8 * len(n_estimatorses)))
    worst_score = scores.min()
    for i, n_estimators in enumerate(n_estimatorses):
        ax = fig.add_subplot(len(n_estimatorses), 1, i + 1)
        im = ax.imshow(
            scores[i,:,:],
            origin="lower",
            vmin=worst_score, vmax=best_score
        )
        ax.set_xticks(
            np.arange(len(max_sampleses)),
            labels=[f"{v:.2f}" for v in max_sampleses]
        )
        ax.set_xlabel("max_samples")
        ax.set_yticks(
            np.arange(len(max_depths)),
            labels=[f"{v}" for v in max_depths]
        )
        ax.set_ylabel("max_depth")
        plt.setp(ax.get_xticklabels(), rotation=90)
        ax.set_title(f"n_estimators = {n_estimators}")
        cbar = fig.colorbar(im, ax=ax, shrink=0.7, label="Area under ROC curve")
    fig.show()

    pl.set_params(**best_params)
    pl.fit(train_data, target_data.loc[train_data.index])
    return best_score, best_params, pl
```

A nyn√≠ zkusme zavolat funkci `find_random_forest` s t√≠m, ≈æe opƒõt vynech√°me v≈°echny p≈ô√≠znaky z `STUFF_TO_DROP` krom `opening_eco`. Pozor, vyhodnocov√°n√≠ bu≈àky n√≠≈æe trv√° nƒõkolik minut.

```python
plt.rcParams.update(plt.rcParamsDefault)
```

```python
%matplotlib inline

random_forest_score, random_forest_params, random_forest_pipeline = find_random_forest([
    prep_add_game_len_secs,
    mk_prep_dropper({"opening_eco"}), # drop all columns from STUFF_TO_DROP except opening_eco
    prep_one_hot_categories,
])

print(f"best score: {random_forest_score:.3f}, best params: {random_forest_params}")
```

Na heatmap√°ch v√Ω≈°e m≈Ø≈æeme pozorovat n√°sleduj√≠c√≠ z√°le≈æitosti:
- ƒç√≠m hlub≈°√≠ jsou v lese stromy, t√≠m l√©p≈°√≠ poskytuje odhady (lze p≈ôedpokl√°dat, ≈æe optim√°ln√≠ hloubka strom≈Ø je 7, tak, jako tomu bylo u samotn√©ho rozhodovac√≠ho stromu)
- √∫spƒõ≈°nost les≈Ø roste relativnƒõ rychle v z√°vislosti na poƒçtu strom≈Ø od `n_estimators = 100` do `n_estimators = 200`, zejm√©na pro ni≈æ≈°√≠ hodnoty `max_samples`
- pro `n_estimators >= 200` z≈ôejmƒõ plat√≠, ≈æe ƒç√≠m n√≠≈æ≈°√≠ `max_samples`, t√≠m vy≈°≈°√≠ √∫spƒõ≈°nost (lze p≈ôedpokl√°dat, ≈æe existuje hodnota `max_samples`, kter√° je dost mal√° na to, aby pro ni≈æ≈°√≠ hodnoty u≈æ √∫spƒõ≈°nost klesala, ale tato hodnota je urƒçitƒõ men≈°√≠ ne≈æ 0.2).

Pov≈°imnƒõme si t√©≈æe, ≈æe √∫spƒõ≈°nost modelu je jen 0.687, narozd√≠l od samotstatn√©ho stromu, kde jsme pro hloubku 7 dos√°hli √∫spƒõ≈°nosti 0.7. Domn√≠v√°me se, ≈æe je to povahou probl√©mu; ≈°achy jsou velice komplexn√≠ z√°le≈æitost, a proto neplat√≠, ≈æe zv√Ω≈°√≠me-li poƒçet strom≈Ø, m≈Ø≈æeme sn√≠≈æit jejich jednotliv√© hloubky a dos√°hnout srovnateln√Ωch v√Ωsledk≈Ø, co≈æ asi pro mnoho jin√Ωch probl√©m≈Ø plat√≠.

Zkusme proto natr√©novat je≈°tƒõ jeden n√°hodn√Ω les, ale tentokr√°t mu neposkytneme ani p≈ô√≠znak `opening_eco`. Nebude m√≠t tedy k dispozici ≈æ√°dn√© informace o jednotliv√Ωch taz√≠ch, pouze obecn√© parametry odehran√© partie.

```python
random_forest_score_2, random_forest_params_2, random_forest_pipeline_2 = find_random_forest([
    prep_add_game_len_secs,
    mk_prep_dropper(), # drop all columns from STUFF_TO_DROP
    prep_one_hot_categories,
])

print(f"best score: {random_forest_score_2}, best params: {random_forest_params_2}")
```

## Vyhodnocen√≠ modelu

K vykreslen√≠ rozhodovac√≠ hranice si vybereme dva nejd≈Øle≈æitƒõj≈°√≠ p≈ô√≠znaky.

```python
best_model.steps[-1][1].feature_importances_
best_model.steps[-1][1].feature_names_in_
best = pd.DataFrame(data=best_model.steps[-1][1].feature_importances_)
best = best.transpose() 
best.columns = best_model.steps[-1][1].feature_names_in_
best = best.melt()
```

```python
palette = sns.color_palette("husl", 4)
sns.set_palette(palette)

fig, ax1 = plt.subplots(figsize=(10,5))
sns.barplot(data = best, x = 'variable', y = 'value', ax = ax1)
fig.suptitle('Feature importance in best model')
ax1.bar_label(ax1.containers[0], fmt='%.3f')

None
```

Vzhledem k vƒõt≈°√≠mu poƒçtu p≈ô√≠znak≈Ø bude pomƒõrnƒõ n√°roƒçn√© podle nich udƒõlat rozhodovac√≠ hranici. Nicm√©nƒõ vid√≠me, ≈æe zdaleka nejd≈Ølƒõ≈æitƒõj≈°√≠ p≈ô√≠znaky jsou `white_rating` a `black_rating`, tud√≠≈æ vykresl√≠me rozhodovac√≠ hranici alespo≈à pro ni.

```python
from sklearn.base import clone
reduced = clone(best_model.steps[-1][1])
```

```python
X = test_data[['white_rating', 'black_rating']]
y = target_data.loc[test_data.index].cat.remove_unused_categories()
reduced.fit(X, y)
```

Nevykresl√≠me v≈°echny body, jeliko≈æ by byl graf nep≈ôehledn√Ω, tud√≠≈æ si vybereme n√°hodnƒõ pƒõtinu bod≈Ø.

```python
X_sample = X.sample(frac=0.2, random_state=1212)
y_sample = y.loc[X_sample.index]
```

```python
from sklearn.inspection import DecisionBoundaryDisplay
fig, ax1 = plt.subplots(figsize=(8, 6))
disp = DecisionBoundaryDisplay.from_estimator(
    reduced, X, response_method="predict",
    xlabel='white_rating', ylabel='black_rating',
    alpha=0.5, ax = ax1)
disp.ax_.scatter(X_sample['white_rating'], X_sample['black_rating'], c=y_sample, edgecolor="k")
fig.suptitle('Prediction boundary')
```

I na na≈°em zjednodu≈°en√©m modelu vid√≠me d≈Øle≈æit√©, aƒçkoliv nep≈ô√≠li≈° p≈ôekvapiv√© v√Ωsledky. V hr√°ch, kde mƒõl znaƒçnƒõ vy≈°≈°√≠ rating b√≠l√Ω/ƒçern√Ω, tak model s velkou pravdƒõpodobnost√≠ vybral jako v√≠tƒõze onoho hr√°ƒçe. Ve hr√°ch s tƒõsn√Ωm rozd√≠lem rating≈Ø u≈æ to bohu≈æel nebylo tak snadn√©.

Nyn√≠ vykresl√≠me dva zaj√≠mav√© indik√°tory √∫spƒõ≈°nosti modelu. Zaƒçnƒõme s matic√≠ z√°mƒõn.

```python
from sklearn.metrics import ConfusionMatrixDisplay

ConfusionMatrixDisplay.from_estimator(
    best_model,
    test_data,
    target_data.loc[test_data.index],
    cmap="Blues",
)
plt.title("Confusion matrix")
plt.grid(False)
plt.show()
```

Z matice z√°mƒõn m≈Ø≈æeme vyƒç√≠st, ≈æe n√°≈° model nen√≠ p≈ô√≠li≈° dobr√Ω v predikci v√Ωhry ƒçern√©ho. Predikce b√≠l√©ho jsou oproti tomu velmi dobr√©. Je mo≈æn√©, ≈æe to je zp≈Øsobeno t√≠m, ≈æe b√≠l√Ω je vƒõt≈°inou ten, kter√Ω vyb√≠r√° zah√°jen√≠ a model jej ƒçastƒõji oznaƒç√≠ jako v√Ωherce dan√©ho zah√°jen√≠.

Nyn√≠ je≈°tƒõ vykresl√≠me ROC k≈ôivku pro n√°≈° model.

```python
from sklearn.metrics import roc_curve, roc_auc_score

probs = best_model.predict_proba(test_data)[:,1]
fpr, tpr, thresholds = roc_curve(target_data.loc[test_data.index], probs, pos_label = 'white')
```

```python

plt.style.use('seaborn-darkgrid')
fig, ax = plt.subplots(figsize=(8,6))
ax.plot(fpr, tpr, label='ROC (AUC = {:.2f})'.format(roc_auc_score(target_data.loc[test_data.index], probs, )))
ax.plot([0,1], [0,1], '--', label='Random')

ax.legend()
ax.set_xlabel('False positive rate')
ax.set_ylabel('True positive rate')

ax.set_title('ROC curve')

None
```
