```python
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec

# Taken from 4th notebook of BI-VIZ course
def univariate_cont_ds(df, column, title):
    stats_df = pd.DataFrame(np.array([
        ["Minimal value", df[column].min()], 
        ["Maximal value", df[column].max()], 
        ["Range", df[column].max() - df[column].min()],
        ["Average", df[column].mean()],
        ["Median", df[column].median()],
        ["Variance", df[column].var()]
    ]), columns=['Title', 'Value'])
    
    display(stats_df)
    
    # figure layout initialization
    fig = plt.figure(figsize=(10,10), constrained_layout=True)
    fig.suptitle(title, fontsize = 18)
    spec = gridspec.GridSpec(ncols=2, nrows=2, figure=fig)
    ax1 = fig.add_subplot(spec[0, :]) 
    ax2 = fig.add_subplot(spec[1, 0]) 
    ax3 = fig.add_subplot(spec[1, 1])
    
    sns.histplot(data=df, x=column, kde=True, ax=ax1)
    
    # x and y - rotation of plots
    sns.boxplot(data=df, y=column, ax=ax2, medianprops={"color": "pink"})
    sns.violinplot(data=df, y=column, ax=ax3)
```

```python
sns.set_style("darkgrid")

palette = sns.color_palette("hls", 10)
sns.set_palette(palette)
```

Nyní by nás mohlo zajímat nejhranější formáty podle počátečních časů a jejich rozdělení do FIDE časových formátů.

**Herní formáty**
- `Bullet`: Do 3 minut
- `Blitz`: 3-10 minut
- `Rapid`: 10-60 minut
- `Classical`: 60 a více minut

```python
fo = orig_data[['increment_code']].copy()
def convert(code):
    return int(code.split('+')[0])
def get_format(init):
    if init < 3:
        return 'Bullet'
    elif init < 10:
        return 'Blitz'
    elif init < 60:
        return 'Rapid'
    else:
        return 'Classical'
    
fo['initial_time'] = fo['increment_code'].apply(convert)
fo['format'] = fo['initial_time'].apply(get_format)
```

```python
fo['initial_time'].replace(pd.value_counts(fo['initial_time']).iloc[8:].index, 'Other', inplace = True)
```

```python
mod = op[(op.winner != 'draw') & (op.op_name.isin(pd.value_counts(op['op_name']).iloc[:15].index))].copy()
mod['winner'] = mod['winner'].cat.remove_unused_categories()
mod['winner'].unique()
mod = mod.groupby('op_name')['winner'].value_counts(normalize=True)
mod = mod.mul(100)
mod = mod.rename('percent').reset_index()
mod.rename(columns={"level_1": "winner"}, inplace = True)
```

```python
from sklearn.model_selection import train_test_split

train_data, test_data = train_test_split(orig_data, test_size=0.25, random_state=SEED) # separate test data
train_data, val_data = train_test_split(train_data, test_size=0.30, random_state=SEED) # separate validation data

print(f"train_data: {len(train_data)}, val_data: {len(val_data)}, test_data: {len(test_data)}")
```

V rámci výběru modelu můžeme experimentovat i s různými dalšími způsoby výběru a transformace příznaků. Tyto úpravy budeme považovat za součást modelu a zabalíme je společně s natrénovaným modelem do Pipeline.

Definujme nyní nějaké pomocné funkce pro úpravu příznaků.

```python
def prep_add_game_len_secs(df):
    df = df.copy()
    df["game_len_s"] = df["last_move_at"] - df["created_at"]
    return df

STUFF_TO_DROP = [
    "victory_status",
    "last_move_at",
    "created_at",
    "moves",
    "increment_code",
    "white_id",
    "black_id",
    "opening_name",
    "opening_eco",
]
def mk_prep_dropper(keep=set()):
    def prep_drop(df):
        df = df.copy()
        for col in STUFF_TO_DROP:
            if col not in keep:
                df.drop(col, inplace=True, axis=1)
        return df
    return prep_drop

def prep_one_hot_categories(df: pd.DataFrame):
    for col in df.columns:
        if df[col].dtype == "category":
            df = df.join(pd.get_dummies(df[col])).drop(col, axis=1)
    return df
```

## Trénování modelu

Nyní definujme funkci, která dostane parametrem funkce pro úpravu příznaků a pokusí se najít optimální rozhodovací strom. Tím je míněno to, že bude zkoušet různé hodnoty hyperparametrů, a vybere ten rozhodovací strom, který je po natrénování na testovacích datech nejúspěšnější na validační datech. Hyperparametry, které budeme ladit, bude `criterion`, aneb kritérium, pomocí něhož se měří informační zisk, a `max_depth`, tedy hloubka stromu.

Úspěšnost modelu budeme měřit plochou pod ROC křivkou. ROC křivka je velmi užitečná metoda pro vizualizaci toho, jak dobře fungují modely pro klasifikaci. ROC křivku vykreslíme níže. Nyní budeme pouze využívat obsah plochy pod ní, což je číslo, pro nějž platí, že čím je vyšší, tím lepší je kvalita modelu. Hodnota 0.5 odpovídá modelu, který činí náhodné predikce, a hodnota 1 odpovídá dokonalému modelu, který má vždy pravdu. Lze proto očekávat, že hodnota pro náš model se bude pohybovat v rozmezí 0.5-1.0. 

Funkce také zobrazí závislost úspěšnosti na validačních datech vůči hloubce rozhodovacího stromu.

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import FunctionTransformer
from sklearn.metrics import roc_auc_score

def find_decision_tree(prep_fns):
    pl = Pipeline(
        [
            (f"fn{i}", FunctionTransformer(fn))
            for i, fn in enumerate(prep_fns)
        ] + [
            ("clf", DecisionTreeClassifier(
                criterion="entropy",
                max_depth=10,
                max_features=None,
                random_state=SEED,
            ))
        ]
    )

    best_score = None
    best_params = None
    
    scores = { "entropy": [], "gini": [] }
    max_depths = np.arange(1, 16)
    for max_depth in max_depths:
        for criterion in ("entropy", "gini"):
            candidate_params = {
                "clf__max_depth": max_depth,
                "clf__criterion": criterion,
            }

            pl.set_params(**candidate_params)
            pl.fit(train_data, target_data.loc[train_data.index])

            proba = pl.predict_proba(val_data)
            candidate_score = roc_auc_score(
                target_data.loc[val_data.index],
                proba if proba.shape[1] > 2 else proba[:, 1],
                multi_class="ovo"
            )

            scores[criterion].append(candidate_score)
            if best_score is None or candidate_score > best_score:
                best_score = candidate_score
                best_params = candidate_params

    fig = plt.figure(figsize=(10,8))
    ax = fig.add_subplot()
    ax.plot(max_depths, scores["entropy"], label="entropy")
    ax.plot(max_depths, scores["gini"], label="gini")
    ax.set_xlabel("max_depth")
    ax.set_ylabel("roc_auc_score")
    ax.legend()
    fig.show()

    pl.set_params(**best_params)
    pl.fit(train_data, target_data.loc[train_data.index])
    return best_score, pl
```

Zavolejme nyní tuto funkci a nakresleme výsledný rozhodovací strom. Zkusme zahodit všechny příznaky uvedené v seznamu `STUFF_TO_DROP` kromě `opening_eco`.

```python
decision_tree_score, decision_tree_pipeline = find_decision_tree([
    prep_add_game_len_secs,
    mk_prep_dropper({"opening_eco"}), # drop all columns from STUFF_TO_DROP except opening_eco
    prep_one_hot_categories,
])

decision_tree: DecisionTreeClassifier = decision_tree_pipeline.steps[-1][1]
print(f"score: {decision_tree_score:.3f}, depth: {decision_tree.max_depth}, criterion: {decision_tree.criterion}, max_features: {decision_tree.max_features}")
```

Jak vidno, zvítězil rozhodovací strom s hloubkou 7 využívající entropii pro měření informačního zisku. Na vizualizaci výše znázorňujeme závislost plochy pod ROC křivkou (metrika, kterou využíváme pro měření úspěšnosti modelu) na hloubce stromu, a to zvlášť pro stromy natrénované pomocí Entropie a stromy natrénované pomocí Gini indexu. Entropie a Gini index zřejme při binární klasifikaci vedou na velmi podobné výsledky (ale předtím jsme zkoušeli ternární klasifikaci a tam Entropie dosahovala viditelně lepších úspěšností, nežli Gini index). Dále pozorujeme následující trend: úspěšnost modelu s rostoucí hloubkou stromu nejdříve roste, a poté klesá. Klesání je způsobeno přeučením - hlubší stromy se více adaptují na trénovací množinu na úkor toho, jak jsou schopni zgeneralizovat svoje znalosti na validační množinu. Věříme proto, že toto klesání pokračuje do nekonečna, a pro hloubku 7 skutečně nastává globální maximum pro Entropii.

Nyní můžeme strom i nakreslit pomocí funkce `plot_tree`, ale v zájmu přehlednosti zobrazíme jen první tři hladiny.

```python
from sklearn.tree import plot_tree

fig = plt.figure(figsize=(16,16))
_ = plot_tree(decision_tree, ax=fig.add_subplot(), max_depth=2, fontsize=12, feature_names=decision_tree.feature_names_in_)
fig.show()
```

Nyní zkusíme problém řešit pomocí náhodných lesů. Hyperparametry, které budeme ladit, budou:
- `n_estimators`, tedy počet stromů v náhodném lese.
- `max_depth`, tedy hloubka jednotlivých stromů v lese
- `max_samples`, tedy velikost náhodného výběru z původních dat, na kterém bude každý strom natrénován (každý strom bude natrénován na jiném náhodném výběru, v tom je kouzlo náhodného lesa).
  
Parametr `criterion` nastavíme na `"entropy"` a parametru `max_features` ponecháme jeho původní hodnotu `"sqrt"`, která by měla dávat dobré výsledky.

Funkce `find_random_forest` opět vykreslí vizualizaci úspěšnosti modelu v závislosti na hyperparametrech, tentokrát ovšem bude zkoumaný prostor hyperparametrů tvaru 5x5x5, takže při vizualizaci použijeme 5 heatmap.

```python
from sklearn.ensemble import RandomForestClassifier

def find_random_forest(prep_fns, **kwargs):
    pl = Pipeline(
        [
            (f"fn{i}", FunctionTransformer(fn))
            for i, fn in enumerate(prep_fns)
        ] + [
            ("clf", RandomForestClassifier(**{
                **{
                    "criterion": "entropy",
                    "max_depth": 1,
                    "max_samples": 0.1,
                    "random_state": SEED,
                    "n_estimators": 1000,
                },
                **kwargs
            }))
        ]
    )

    best_score = None
    best_params = None

    n_estimatorses = np.arange(100, 501, 100)
    max_depths = np.arange(1, 6)
    max_sampleses = np.linspace(0.2, 1, 5)
    scores = np.zeros((
        len(n_estimatorses),
        len(max_depths),
        len(max_sampleses)
    ))
    
    for i, n_estimators in enumerate(n_estimatorses):
        print(f"level {i + 1}/{len(n_estimatorses)}: ", end="")
        for j, max_depth in enumerate(max_depths):
            print("|", end="")
            for k, max_samples in enumerate(max_sampleses):
                print(".", end="")
                candidate_params = {
                    "clf__n_estimators": n_estimators,
                    "clf__max_depth": max_depth,
                    "clf__max_samples": max_samples,
                }

                pl.set_params(**candidate_params)
                pl.fit(train_data, target_data.loc[train_data.index])

                proba = pl.predict_proba(val_data)
                candidate_score = roc_auc_score(
                    target_data.loc[val_data.index],
                    proba if proba.shape[1] > 2 else proba[:, 1],
                    multi_class="ovo"
                )

                scores[i, j, k] = candidate_score
                if best_score is None or candidate_score > best_score:
                    best_score = candidate_score
                    best_params = candidate_params
        print()

    fig = plt.figure(figsize=(8,8 * len(n_estimatorses)))
    worst_score = scores.min()
    for i, n_estimators in enumerate(n_estimatorses):
        ax = fig.add_subplot(len(n_estimatorses), 1, i + 1)
        im = ax.imshow(
            scores[i,:,:],
            origin="lower",
            vmin=worst_score, vmax=best_score
        )
        ax.set_xticks(
            np.arange(len(max_sampleses)),
            labels=[f"{v:.2f}" for v in max_sampleses]
        )
        ax.set_xlabel("max_samples")
        ax.set_yticks(
            np.arange(len(max_depths)),
            labels=[f"{v}" for v in max_depths]
        )
        ax.set_ylabel("max_depth")
        plt.setp(ax.get_xticklabels(), rotation=90)
        ax.set_title(f"n_estimators = {n_estimators}")
        cbar = fig.colorbar(im, ax=ax, shrink=0.7, label="Area under ROC curve")
    fig.show()

    pl.set_params(**best_params)
    pl.fit(train_data, target_data.loc[train_data.index])
    return best_score, best_params, pl
```

A nyní zkusme zavolat funkci `find_random_forest` s tím, že opět vynecháme všechny příznaky z `STUFF_TO_DROP` krom `opening_eco`. Pozor, vyhodnocování buňky níže trvá několik minut.

```python
plt.rcParams.update(plt.rcParamsDefault)
```

```python
%matplotlib inline

random_forest_score, random_forest_params, random_forest_pipeline = find_random_forest([
    prep_add_game_len_secs,
    mk_prep_dropper({"opening_eco"}), # drop all columns from STUFF_TO_DROP except opening_eco
    prep_one_hot_categories,
])

print(f"best score: {random_forest_score:.3f}, best params: {random_forest_params}")
```

Na heatmapách výše můžeme pozorovat následující záležitosti:
- čím hlubší jsou v lese stromy, tím lépší poskytuje odhady (lze předpokládat, že optimální hloubka stromů je 7, tak, jako tomu bylo u samotného rozhodovacího stromu)
- úspěšnost lesů roste relativně rychle v závislosti na počtu stromů od `n_estimators = 100` do `n_estimators = 200`, zejména pro nižší hodnoty `max_samples`
- pro `n_estimators >= 200` zřejmě platí, že čím nížší `max_samples`, tím vyšší úspěšnost (lze předpokládat, že existuje hodnota `max_samples`, která je dost malá na to, aby pro nižší hodnoty už úspěšnost klesala, ale tato hodnota je určitě menší než 0.2).

Povšimněme si téže, že úspěšnost modelu je jen 0.687, narozdíl od samotstatného stromu, kde jsme pro hloubku 7 dosáhli úspěšnosti 0.7. Domníváme se, že je to povahou problému; šachy jsou velice komplexní záležitost, a proto neplatí, že zvýšíme-li počet stromů, můžeme snížit jejich jednotlivé hloubky a dosáhnout srovnatelných výsledků, což asi pro mnoho jiných problémů platí.

Zkusme proto natrénovat ještě jeden náhodný les, ale tentokrát mu neposkytneme ani příznak `opening_eco`. Nebude mít tedy k dispozici žádné informace o jednotlivých tazích, pouze obecné parametry odehrané partie.
